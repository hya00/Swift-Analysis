```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
swift_xrt = pd.read_table('grb_xrt.txt', sep='\t')
swift_xrt = swift_xrt.rename(columns = {'GRB': 'GRB Name'})
swift_xrt['GRB Name'] = 'GRB' + swift_xrt['GRB Name']
swift_xrt['XRT 24 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'] = swift_xrt['XRT 24 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'].astype(float)
swift_xrt['XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'] = swift_xrt['XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'].astype(float)
```

```python
swift_grbs = pd.read_csv('swift_data.csv', sep='\t')
```

```python
swift_xrt_grbs = pd.merge(swift_grbs, swift_xrt)

LGRBs = swift_xrt_grbs.loc[swift_xrt_grbs['T90']>2]
SGRBs = swift_xrt_grbs.loc[swift_xrt_grbs['T90']<2]

LGRBs_lowz = LGRBs.loc[LGRBs['z']<1]
LGRBs_highz  = LGRBs.loc[LGRBs['z']>1]
SGRBs_lowz = SGRBs.loc[SGRBs['z']<1]
SGRBs_highz  = SGRBs.loc[SGRBs['z']>1]
```

```python
plt.figure(figsize=(20,10)) #sets figure size
ax = plt.gca()
ax.set_xscale('log')

plt.scatter(LGRBs_lowz['T90'], LGRBs_lowz['XRT Spectral Index (Gamma)'])
plt.scatter(LGRBs_highz['T90'], LGRBs_highz['XRT Spectral Index (Gamma)'])
plt.scatter(SGRBs_lowz['T90'], SGRBs_lowz['XRT Spectral Index (Gamma)'])
plt.scatter(SGRBs_highz['T90'], SGRBs_highz['XRT Spectral Index (Gamma)'])

plt.xlabel('T90')
plt.ylabel('Spectral Index (Gamma)')
plt.legend(['LGRBs low z', 'LGRBs high z','SGRBs low z', 'SGRBs high z'])
plt.show()
```

```python
plt.figure(figsize=(20,10)) #sets figure size
ax = plt.gca()

plt.scatter(LGRBs['z'], LGRBs['XRT Spectral Index (Gamma)'])
plt.scatter(SGRBs['z'], SGRBs['XRT Spectral Index (Gamma)'])

plt.xlabel('z')
plt.ylabel('Spectral Index (Gamma)')
plt.legend(['LGRBs','SGRBs'])
plt.show()
```

```python
plt.figure(figsize=(20,10)) #sets figure size

n, bins, patches = plt.hist(SGRBs_lowz['XRT Spectral Index (Gamma)'], len(SGRBs_lowz), density=True, histtype='step',
                         label='Empirical')
m, bins, patches = plt.hist(LGRBs_lowz['XRT Spectral Index (Gamma)'], len(LGRBs_lowz), density=True, histtype='step'
                            , label='Empirical')
plt.xlabel('spectral index')
plt.ylabel('No. of GRBs')
plt.legend(['SGRBs low z', 'LGRBs low z'])
plt.show()
```

```python
plt.figure(figsize=(20,10)) #sets figure size
ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')

plt.scatter(LGRBs_lowz['T90'], LGRBs_lowz['XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'])
plt.scatter(LGRBs_highz['T90'], LGRBs_highz['XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'])
plt.scatter(SGRBs_lowz['T90'], SGRBs_lowz['XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'])
plt.scatter(SGRBs_highz['T90'], SGRBs_highz['XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]'])

plt.xlabel('T90')
plt.ylabel('XRT 11 Hour Flux (0.3-10 keV) [10^-11 erg/cm^2/s]')
plt.legend(['LGRBs low z', 'LGRBs high z','SGRBs low z', 'SGRBs high z'])
plt.show()
```
