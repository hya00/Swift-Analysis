```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
swift_xrt_grbs = pd.read_csv('swift_xrt_grbs')
LGRBs = swift_xrt_grbs.loc[swift_xrt_grbs['T90']>2]
SGRBs = swift_xrt_grbs.loc[swift_xrt_grbs['T90']<2]

LGRBs_lowz = LGRBs.loc[LGRBs['z']<1]
LGRBs_highz  = LGRBs.loc[LGRBs['z']>1]
SGRBs_lowz = SGRBs.loc[SGRBs['z']<1]
SGRBs_highz  = SGRBs.loc[SGRBs['z']>1]
```

```python
plt.figure(figsize=(15,8))

ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')

plt.scatter(LGRBs_lowz['T90'], LGRBs_lowz['XRT Column Density (NH) [10^21 cm^-2]'])
plt.scatter(LGRBs_highz['T90'], LGRBs_highz['XRT Column Density (NH) [10^21 cm^-2]'])
plt.scatter(SGRBs_lowz['T90'], SGRBs_lowz['XRT Column Density (NH) [10^21 cm^-2]'])
plt.scatter(SGRBs_highz['T90'], SGRBs_highz['XRT Column Density (NH) [10^21 cm^-2]'])

plt.xlabel('T90')
plt.ylabel('XRT Column Density (NH) [10^21 cm^-2]')
plt.legend(['LGRBs low z', 'LGRBs high z','SGRBs low z', 'SGRBs high z'])
plt.show()
```

```python
plt.figure(figsize=(20,10)) #sets figure size
ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
plt.scatter(LGRBs['z']+1, LGRBs['XRT Column Density (NH) [10^21 cm^-2]'])
plt.scatter(SGRBs['z']+1, SGRBs['XRT Column Density (NH) [10^21 cm^-2]'])

plt.xlabel('z')
plt.ylabel('XRT Column Density (NH) [10^21 cm^-2]')
plt.legend(['LGRBs','SGRBs'])
plt.show()
```
